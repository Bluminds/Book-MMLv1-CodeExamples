### EJERCICIOS PRÁCTICO RESUELTOS DEL CAPÍTULO

```python
# Ejercicio 1
# Variables de diferentes tipos
entero = 10
flotante = 3.14
cadena = "Hola Mundo"
booleano = True

# Mostrar tipos y valores
print(type(entero), entero)
print(type(flotante), flotante)
print(type(cadena), cadena)
print(type(booleano), booleano)
```

```python
# Ejercicio 2
# Números para operaciones
a = 15
b = 5

# Operaciones aritméticas
suma = a + b
resta = a - b
multiplicacion = a * b
division = a / b

# Mostrar resultados
print("Suma:", suma)
print("Resta:", resta)
print("Multiplicación:", multiplicacion)
print("División:", division)
```

```python
# Ejercicio 3
# Cadenas para concatenar
cadena1 = "Hola"
cadena2 = "Mundo"

# Concatenación
resultado = cadena1 + " " + cadena2

# Mostrar resultado
print(resultado)
```

```python
# Ejercicio 4
# Lista de números
numeros = [1, 2, 3, 4]

# Añadir un número al final
numeros.append(5)

# Mostrar lista modificada
print(numeros)
```

```python
# Ejercicio 5
# Número para evaluar
numero = -3

# Condicionales
if numero > 0:
    print("Positivo")
elif numero < 0:
    print("Negativo")
else:
    print("Cero")
```

```python
# Ejercicio 6
# Número entero y cadena
numero_entero = 7
cadena_numero = "42"

# Conversión de tipos
numero_flotante = float(numero_entero)
numero_entero_convertido = int(cadena_numero)

# Mostrar resultados
print(numero_flotante)
print(numero_entero_convertido)
```

```python
# Ejercicio 7
# Lista de frutas
frutas = ["manzana", "plátano", "cereza", "fresa"]

# Acceder al tercer elemento
tercera_fruta = frutas[2]

# Mostrar resultado
print(tercera_fruta)
```

```python
# Ejercicio 8
# Números para comparar
a = 10
b = 15

# Comparaciones
son_iguales = a == b
a_es_mayor = a > b

# Mostrar resultados
print("Son iguales:", son_iguales)
print("A es mayor que B:", a_es_mayor)
```

```python
# Ejercicio 9
# # Diccionario con datos del libro
libro = {
    "título": "Cien años de soledad",
    "autor": "Gabriel García Márquez",
    "año": 1967,
}

# Mostrar el año
print(libro["año"])
```

```python
# Ejercicio 10
# Definir función
def sumar(a, b):
    return a + b


# Llamar a la función
resultado = sumar(5, 3)

# Mostrar resultado
print(resultado)
```

```python
# Ejercicio 11
# Lista de números
numeros = [10, 20, 30, 40, 50]

# Obtener sublista
sublista = numeros[1:4]

# Mostrar sublista
print(sublista)
```

```python
# Ejercicio 12
# Lista de nombres
nombres = ["Ana", "Juan", "Carlos"]

# Iterar e imprimir cada nombre
for nombre in nombres:
    print(nombre)
```

```python
# Ejercicio 13
# Contador inicial
i = 1

# Bucle while
while i <= 5:
    print(i)
    i += 1
```

```python
# Ejercicio 14
# Definir la función
def saludar(nombre, saludo="Hola"):
    return f"{saludo}, {nombre}!"


# Llamar a la función
print(saludar("Marta"))
print(saludar("Juan", "Buenos días"))
```

```python
# Ejercicio 15
# Números para dividir
a = 10
b = 0

# Intento de división
try:
    resultado = a / b
    print("Resultado:", resultado)
except ZeroDivisionError:
    print("Error: No se puede dividir por cero.")
```

```python
# Ejercicio 16
numero_entero = 7
numero_flotante = 3.5

suma = numero_entero + numero_flotante
resta = numero_entero - numero_flotante
multiplicacion = numero_entero * numero_flotante
division = numero_entero / numero_flotante

print("Suma", suma)
print("Resta", suma)
print("Multiplicación", suma)
print("División", suma)
```
