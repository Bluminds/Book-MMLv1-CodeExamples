```python
# Ejercicio 10
import numpy as np
import matplotlib.pyplot as plt

# Crear una matriz cuadrada diagonalizable
B = np.array([[5, 2, 8], [2, 3, 3], [7, 9, 1]])

# Calculando eigenvalores y eigenvectores
eigenvalores, eigenvectores = np.linalg.eig(B)

# Mostrar eigenvalores y eigenvectores
print("Eigenvalores:")
print(eigenvalores)
print("\nEigenvectores:")
print(eigenvectores)

# Diagonalización: B = PDP^(-1)
D = np.diag(eigenvalores)
P = eigenvectores
P_inv = np.linalg.inv(P)
B_diagonalizada = P @ D @ P_inv

# Visualización de la matriz original y su versión diagonalizada
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Matriz original
ax[0].matshow(B, cmap="viridis")
ax[0].set_title("Matriz Original B")
for (i, j), val in np.ndenumerate(B):
    ax[0].text(j, i, f"{val:.0f}", ha="center", va="center", color="white")

# Matriz diagonalizada
ax[1].matshow(B_diagonalizada, cmap="viridis")
ax[1].set_title("B Diagonalizada")
for (i, j), val in np.ndenumerate(B_diagonalizada):
    ax[1].text(j, i, f"{val:.2f}", ha="center", va="center", color="white")
plt.show()
```

```python
# Ejercicio 11
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy import stats

# Generando un conjunto de datos que cumpla con las condiciones del Teorema de Gauss-Markov
np.random.seed(0)
X = 2.5 * np.random.rand(30, 1)  # Variables independientes
epsilon = np.random.randn(30, 1) * 0.5  # Error normalmente distribuido
Y = 3 + 2 * X + epsilon  # Modelo lineal

# Aplicando MCO (Mínimos Cuadrados Ordinarios)
model_GM = LinearRegression()
model_GM.fit(X, Y)
Y_pred_GM = model_GM.predict(X)

# Cálculo de intervalos de confianza para los parámetros del modelo
X_with_intercept = np.hstack([np.ones((X.shape[0], 1)), X])
hat_matrix = (
    X_with_intercept
    @ np.linalg.inv(X_with_intercept.T @ X_with_intercept)
    @ X_with_intercept.T
)
residuals = Y - Y_pred_GM
variance = (residuals.T @ residuals) / (
    X_with_intercept.shape[0] - X_with_intercept.shape[1]
)
std_error = np.sqrt(
    np.diag(variance[0, 0] * np.linalg.inv(X_with_intercept.T @ X_with_intercept))
)

# Intervalos de confianza para las predicciones
t_value = stats.t.ppf(0.975, df=len(X) - 2)
y_standard_error = np.sqrt(variance[0, 0]) * np.sqrt(
    1 / len(X) + (X - np.mean(X)) ** 2 / np.sum((X - np.mean(X)) ** 2)
)
ci_lower = Y_pred_GM - t_value * y_standard_error
ci_upper = Y_pred_GM + t_value * y_standard_error

# Visualización con intervalo de confianza
plt.scatter(X, Y, color="blue", label="Datos Observados", marker="d")
plt.plot(X, Y_pred_GM, color="red", label="Línea de Regresión MCO")
plt.fill_between(
    X.flatten(),
    ci_lower.flatten(),
    ci_upper.flatten(),
    color="red",
    alpha=0.2,
    label="Intervalo de Confianza 95%",
)
plt.title("Regresión Lineal con Intervalos de Confianza")
plt.xlabel("Variable Independiente")
plt.ylabel("Variable Dependiente")
plt.legend()
plt.show()
```

```python
# Ejercicio 16
import numpy as np
import matplotlib.pyplot as plt

# Función de coste (ejemplo simple)
def cost_function(x):
    return x**2 + 5 * np.sin(x)

# Derivada de la función de coste
def derivative(x):
    return 2 * x + 5 * np.cos(x)

# Gradiente descendente
def gradient_descent(starting_point, learning_rate, iterations):
    x = starting_point
    for _ in range(iterations):
        x -= learning_rate * derivative(x)
    return x

# Parámetros para el gradiente descendente
starting_point = 0
learning_rate = 0.01
iterations = 1000

# Aplicar el gradiente descendente
min_point = gradient_descent(starting_point, learning_rate, iterations)

# Visualización de la función de coste y el punto mínimo encontrado
x = np.linspace(-6, 6, 400)
y = cost_function(x)
plt.plot(x, y, label="Función de Coste")
plt.scatter(min_point, cost_function(min_point), color="red", label="Mínimo Encontrado")
plt.title("Optimización con Gradiente Descendente")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
min_point, cost_function(min_point)
```

```python
# Ejercicio 17
import matplotlib.pyplot as plt

# Redefiniendo la función de activación lineal y su derivada
def linear_activation(x):
    return x

def derivative_linear_activation(x):
    return 1

# Pesos iniciales y parámetros
weight = 0.6
input_data = 1.5  # Ejemplo de entrada
expected_output = 2  # Ejemplo de salida esperada
learning_rate = 0.01
num_iterations = 1000

# Almacenar el historial de pesos y salidas para la visualización
weights_history = []
outputs_history = []

# Ajuste del peso usando el gradiente en cada iteración
for _ in range(num_iterations):
    output = linear_activation(weight * input_data)
    error = expected_output - output
    weight += learning_rate * error * derivative_linear_activation(input_data)

    # Guardar el historial
    weights_history.append(weight)
    outputs_history.append(output)

# Visualización del proceso de ajuste
iterations = range(num_iterations)
plt.figure(figsize=(12, 5))

# Visualización del ajuste de pesos
plt.subplot(1, 2, 1)
plt.plot(iterations, weights_history, color="purple")
plt.title("Ajuste de Peso a lo largo de las Iteraciones")
plt.xlabel("Iteración")
plt.ylabel("Peso")

# Visualización de la salida de la red en comparación con la salida esperada
plt.subplot(1, 2, 2)
plt.plot(iterations, outputs_history, label="Salida de la Red Neuronal", color="red")
plt.hlines(
    expected_output,
    0,
    num_iterations,
    colors="blue",
    label="Salida Esperada",
    linestyles="dashed",
)
plt.title("Salida de la Red Neuronal a lo largo de las Iteraciones")
plt.xlabel("Iteración")
plt.ylabel("Salida")
plt.legend()
plt.tight_layout()
plt.show()
weight, linear_activation(weight * input_data)
```

```python
# Ejercicio 22
# Función de coste y su derivada (ejemplo: función cuadrática)
def quadratic_cost_function(x):
    return x**2 + 10 * x
def derivative_quadratic_cost_function(x):
    return 2 * x + 10
def gradient_descent_quadratic(starting_point, learning_rate, iterations):
    x = starting_point
    history = []  # Guardar el historial para la visualización
    for _ in range(iterations):
        x -= learning_rate * derivative_quadratic_cost_function(x)
        history.append(x)
    return x, history

# Parámetros para el gradiente descendente
starting_point_quadratic = -15
learning_rate_quadratic = 0.1
iterations_quadratic = 25

# Aplicar el gradiente descendente
min_point_quadratic, history_quadratic = gradient_descent_quadratic(
    starting_point_quadratic, learning_rate_quadratic, iterations_quadratic
)

# Visualización
x_values_quadratic = np.linspace(-20, 5, 400)
y_values_quadratic = quadratic_cost_function(x_values_quadratic)
plt.plot(x_values_quadratic, y_values_quadratic, label="Función de Coste")
plt.scatter(
    history_quadratic,
    quadratic_cost_function(np.array(history_quadratic)),
    color="red",
    marker="x",
    label="Iteraciones de Gradiente Descendente",
)
plt.title("Optimización con Gradiente Descendente")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
min_point_quadratic
```

```python
# Ejercicio 26
import matplotlib.pyplot as plt
import numpy as np
def continuous_function(x):
    return np.sin(x) + np.cos(1.5 * x)

# Intervalo cerrado
a, b = 0, 4 * np.pi

# Valores en los extremos del intervalo
value_a = continuous_function(a)
value_b = continuous_function(b)

# Elegir un valor intermedio
intermediate_value = (value_a + value_b) / 2

# Encontrar un punto en el intervalo donde la función toma el valor intermedio
from scipy.optimize import fsolve

def function_minus_value(x, value):
    return continuous_function(x) - value

(x_intermediate,) = fsolve(
    function_minus_value, (a + b) / 2, args=(intermediate_value,)
)

# Visualización
x_values = np.linspace(a, b, 400)
y_values = continuous_function(x_values)
plt.plot(x_values, y_values, label="Función Continua")
plt.scatter([a, b], [value_a, value_b], color="red", label="Valores en los Extremos")
plt.scatter(x_intermediate, intermediate_value, color="black", label="Valor Intermedio")
plt.axhline(y=intermediate_value, color="gray", linestyle="--")
plt.title("Teorema del Valor Intermedio")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
a, value_a, b, value_b, x_intermediate, intermediate_value
```

```python
# Ejercicio 27
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar
def continuous_function_example(x):
    return np.sin(x) + 0.2 * np.cos(3 * x) + np.sin(0.6 * x)

# Intervalo cerrado
interval = (0, 2 * np.pi)

# Buscar el valor máximo y mínimo en el intervalo cerrado
min_result = minimize_scalar(
    lambda x: -continuous_function_example(x), bounds=interval, method="bounded"
)
max_result = minimize_scalar(
    continuous_function_example, bounds=interval, method="bounded"
)

# Valores extremos y sus puntos
min_value = continuous_function_example(min_result.x)
max_value = continuous_function_example(max_result.x)

# Visualización
x_values_extreme = np.linspace(interval[0], interval[1], 400)
y_values_extreme = continuous_function_example(x_values_extreme)
plt.plot(x_values_extreme, y_values_extreme, label="Función Continua")
plt.scatter(min_result.x, min_value, color="green", label="Valor Mínimo")
plt.scatter(max_result.x, max_value, color="red", label="Valor Máximo")
plt.title("Teorema del Valor Extremo")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
min_result.x, min_value, max_result.x, max_value
```

```python
# Ejercicio 28
import numpy as np
import matplotlib.pyplot as plt
import numdifftools as nd
def numerator_function(x):
    return np.sin(x) - x
def denominator_function(x):
    return x**2

# Construir los objetos de derivada de orden 1 y 2
d1_num = nd.Derivative(numerator_function, n=1, order=5)
d1_den = nd.Derivative(denominator_function, n=1, order=5)

# Primera aplicación de L'Hopital
num1 = d1_num(0.0)
den1 = d1_den(0.0)

# Como da 0/0, aplicamos L'Hopital de nuevo (segunda derivada)
d2_num = nd.Derivative(numerator_function, n=2, order=5)
d2_den = nd.Derivative(denominator_function, n=2, order=5)

num2 = d2_num(0.0)
den2 = d2_den(0.0)

limit_lhopital = num2 / den2

print(f"1ra derivada: num={num1:.3e}, den={den1:.3e}")
print(f"2da derivada: num={num2:.6f}, den={den2:.6f}")
print(f"Límite (dos veces) -> {limit_lhopital:.6f}")

# Visualización
x_vals = np.linspace(-2, 2, 400)
y_num = numerator_function(x_vals)
y_den = denominator_function(x_vals)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_num, label="numerador: sin(x) - x")
plt.plot(x_vals, y_den, label="denominador: x^2", color="orange")
plt.scatter(0, num2, color="red",   label="2da deriv. num. en 0")
plt.scatter(0, den2, color="green", label="2da deriv. den. en 0")
plt.title("Regla de L'Hopital aplicado dos veces (Numdifftools)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ejercicio 29
from scipy.optimize import fsolve
def function_for_tvm(x):
    return np.sin(x) + 0.5 * x

# Calcular el cambio promedio en el intervalo
a_tvm, b_tvm = 0, np.pi
change_in_function = function_for_tvm(b_tvm) - function_for_tvm(a_tvm)
change_in_variable = b_tvm - a_tvm
average_rate_of_change = change_in_function / change_in_variable

# Encontrar un punto donde la derivada es igual al cambio promedio
def derivative_of_function(x):
    return np.cos(x) + 0.5


# Buscar el punto en el intervalo [a, b] donde la derivada es igual al cambio promedio
(point_tvm,) = fsolve(
    lambda x: derivative_of_function(x) - average_rate_of_change, (a_tvm + b_tvm) / 2
)

# Visualización
x_values_tvm = np.linspace(a_tvm - 1, b_tvm + 1, 400)
y_values_tvm = function_for_tvm(x_values_tvm)
tangent_line = derivative_of_function(point_tvm) * (
    x_values_tvm - point_tvm
) + function_for_tvm(point_tvm)
plt.plot(x_values_tvm, y_values_tvm, label="Función")
plt.plot(
    x_values_tvm,
    tangent_line,
    label="Tangente en el punto del Teorema del Valor Medio",
    linestyle="--",
)
plt.scatter(
    point_tvm,
    function_for_tvm(point_tvm),
    color="red",
    label="Punto del Teorema del Valor Medio",
)
plt.title("Teorema del Valor Medio")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
point_tvm, derivative_of_function(point_tvm), average_rate_of_change
```

```python
# Ejercicio 30
import numdifftools as nd
import numpy as np
import matplotlib.pyplot as plt
import math

# Definir una función y su aproximación de Taylor alrededor de un punto
def original_function(x):
    return np.sin(x)
def taylor_approximation(x, around_point, order):
    taylor_poly = 0
    for n in range(order + 1):
        # Usar numdifftools para calcular las derivadas
        derivative_func = nd.Derivative(original_function, n=n)
        derivative_value = derivative_func(around_point)
        taylor_poly += (derivative_value / math.factorial(n)) * (x - around_point) ** n
    return taylor_poly

# Punto alrededor del cual se calculará la aproximación de Taylor
around_point = 0

# Orden de la aproximación de Taylor
order = 5

# Visualización de la función original y su aproximación de Taylor
x_values_taylor = np.linspace(-2 * np.pi, 2 * np.pi, 400)
y_values_original = original_function(x_values_taylor)
y_values_taylor = taylor_approximation(x_values_taylor, around_point, order)
plt.plot(x_values_taylor, y_values_original, label="Función Original (sin(x))")
plt.plot(
    x_values_taylor,
    y_values_taylor,
    label=f"Aproximación de Taylor de orden {order}",
    linestyle="--",
)
plt.scatter(
    around_point,
    original_function(around_point),
    color="red",
    label="Punto de Aproximación",
)
plt.title("Teorema de Taylor")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
# Ejercicio 31
from scipy.integrate import dblquad
import numpy as np

# Definir un campo vectorial simple (por ejemplo, F(x, y, z) = [x, y, z])
def vector_field(x, y, z):
    return np.array([x, y, z])

# Calcular la divergencia de este campo vectorial
# Divergencia de [x, y, z] es 1 + 1 + 1 = 3
divergence = 3

# Establecer los límites del volumen para la integral de volumen (por ejemplo, un cubo unitario)
volume_limits = (0, 1)  # Límites para x, y, z

# Calcular la integral de volumen de la divergencia
volume_integral, _ = dblquad(
    lambda y, x: divergence,
    volume_limits[0],
    volume_limits[1],
    lambda x: volume_limits[0],
    lambda x: volume_limits[1],
)

# La integral de superficie es más complicada de calcular directamente, pero según el Teorema de la Divergencia,
# debería ser igual a la integral de volumen de la divergencia para una superficie cerrada que encierra el volumen.
# Visualización conceptual (no una representación directa del cálculo)
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Crear una cuadrícula y calcular los componentes del campo vectorial
x, y, z = np.meshgrid(
    np.linspace(volume_limits[0], volume_limits[1], 5),
    np.linspace(volume_limits[0], volume_limits[1], 5),
    np.linspace(volume_limits[0], volume_limits[1], 5),
)
u, v, w = vector_field(x, y, z)

# Dibujar el campo vectorial
ax.quiver(x, y, z, u, v, w, length=0.1)
ax.set_title("Campo Vectorial y Divergencia")
ax.view_init(elev=20, azim=5)
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
plt.show()
volume_integral
```

```python
# Ejercicio 33
from scipy.integrate import odeint

# Definir la ecuación diferencial del sistema (por ejemplo, un amortiguador de resorte)
def spring_damper_system(y, t, k, b):
    # y[0] es la posición y y[1] es la velocidad
    position, velocity = y
    dydt = [velocity, -b * velocity - k * position]
    return dydt

# Constantes del sistema: k (constante de resorte) y b (coeficiente de amortiguamiento)
k = 2.0  # constante de resorte
b = 0.5  # coeficiente de amortiguamiento

# Condiciones iniciales: posición inicial y velocidad inicial
y_initial = [1.0, 0.0]  # [posición inicial, velocidad inicial]

# Crear un arreglo de tiempos para la solución
time = np.linspace(0, 10, 1000)  # 10 segundos

# Resolver la ecuación diferencial
solution = odeint(spring_damper_system, y_initial, time, args=(k, b))

# Visualización de la solución
plt.plot(time, solution[:, 0], label="Posición")
plt.plot(time, solution[:, 1], label="Velocidad", linestyle="--")
plt.title("Sistema de Amortiguador de Resorte")
plt.xlabel("Tiempo")
plt.ylabel("Posición y Velocidad")
plt.legend()
plt.show()
```

```python
# Ejercicio 34
# Importar las librerías necesarias
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definir la dinámica de una neurona en una red neuronal
def neuron_dynamics(state, t):
    # Un modelo sencillo donde la tasa de cambio del estado de la neurona depende de su estado actual
    # y de la entrada recibida. Aquí, usamos una función sigmoide como función de activación.
    return -state + 1 / (1 + np.exp(-state))

# Estado inicial de la neurona
initial_state = 0.1

# Tiempos para la simulación
time = np.linspace(0, 10, 100)

# Resolver la ecuación diferencial para modelar el comportamiento de la neurona
state_over_time = odeint(neuron_dynamics, initial_state, time)

# Visualización de la evolución del estado de la neurona
plt.plot(time, state_over_time)
plt.title("Dinámica de una Neurona en una Red Neuronal")
plt.xlabel("Tiempo")
plt.ylabel("Estado de la Neurona")
plt.show()
```

```python
# Ejercicio 35
# Definir la dinámica de crecimiento de la población
def population_growth(population, t, growth_rate):
    # Modelo simple de crecimiento poblacional: tasa de crecimiento proporcional a la población actual
    dPdt = growth_rate * population
    return dPdt

# Tasa de crecimiento de la población
growth_rate = 0.5

# Población inicial
initial_population = 10

# Resolver la ecuación diferencial para modelar el crecimiento de la población
population_over_time = odeint(
    population_growth, initial_population, time, args=(growth_rate,)
)

# Visualización del crecimiento de la población
plt.plot(time, population_over_time)
plt.title("Crecimiento de una Población a lo Largo del Tiempo")
plt.xlabel("Tiempo")
plt.ylabel("Población")
plt.show()
```

```python
# Ejercicio 36
# Definir la dinámica de un sistema que se adapta a un entorno cambiante
def adaptive_system(state, t, adaptation_rate):
    # Modelo simple donde la tasa de cambio del estado del sistema depende de la tasa de adaptación
    # y del estado del entorno (simulado aquí como una función del tiempo)
    environment_state = np.sin(
        t
    )  # Simulación simplificada del estado cambiante del entorno
    dSdt = adaptation_rate * (environment_state - state)
    return dSdt

# Tasa de adaptación del sistema
adaptation_rate = 0.8

# Estado inicial del sistema
initial_state = 0

# Resolver la ecuación diferencial para modelar el comportamiento adaptativo del sistema
state_over_time = odeint(adaptive_system, initial_state, time, args=(adaptation_rate,))

# Visualización de la adaptación del sistema en el tiempo
plt.plot(time, state_over_time, label="Estado del Sistema")
plt.plot(time, np.sin(time), label="Estado del Entorno", linestyle="--")
plt.title("Adaptación de un Sistema en un Entorno Dinámico")
plt.xlabel("Tiempo")
plt.ylabel("Estado")
plt.legend()
plt.show()
```

```python
# Ejercicio 37
# Simular la optimización de hiperparámetros de un modelo de ML
def hyperparameter_optimization(hyperparameters, t, learning_rate):
    # En este ejemplo simple, asumimos que la optimización se dirige a minimizar alguna función de pérdida
    # Los hiperparámetros se ajustan basados en su efecto sobre la función de pérdida
    dhdt = -learning_rate * np.array(
        [np.sin(h) for h in hyperparameters]
    )  # Simulación simplificada
    return dhdt

# Tasa de aprendizaje para la optimización
learning_rate_hp = 0.1

# Hiperparámetros iniciales
initial_hyperparameters = [0.5, -0.3]  # Ejemplo con dos hiperparámetros

# Resolver la ecuación diferencial para la optimización de hiperparámetros
hyperparameters_over_time = odeint(
    hyperparameter_optimization, initial_hyperparameters, time, args=(learning_rate_hp,)
)

# Visualización de la optimización de hiperparámetros
plt.plot(time, hyperparameters_over_time[:, 0], label="Hiperparámetro 1")
plt.plot(time, hyperparameters_over_time[:, 1], label="Hiperparámetro 2")
plt.title("Optimización de Hiperparámetros en ML")
plt.xlabel("Tiempo")
plt.ylabel("Valor de Hiperparámetros")
plt.legend()
plt.show()
```

```python
# Ejercicio 38
# Modelar una serie temporal compleja y su descomposición usando ecuaciones diferenciales
def time_series_decomposition(state, t, decomposition_rate):
    # 'state' representa los componentes de la serie temporal (tendencia, estacionalidad, etc.)
    # La descomposición se modela como una dinámica que separa estos componentes
    trend = state[0]
    seasonal = np.sin(
        2 * np.pi * t / decomposition_rate
    )  # Simulación de un componente estacional
    return [trend + seasonal, seasonal]

# Tasa de descomposición (periodo estacional)
decomposition_rate_ts = 12

# Estado inicial (componentes de la serie temporal)
initial_state_ts = [0, 0]  # Tendencia inicial y componente estacional

# Resolver la ecuación diferencial para la descomposición de la serie temporal
time_series_solution = odeint(
    time_series_decomposition, initial_state_ts, time, args=(decomposition_rate_ts,)
)

# Visualización de la serie temporal y sus componentes descompuestos
plt.plot(time, time_series_solution[:, 0], label="Serie Temporal Compleja")
plt.plot(
    time, time_series_solution[:, 1], label="Componente Estacional", linestyle="--"
)
plt.title("Descomposición de Series Temporales")
plt.xlabel("Tiempo")
plt.ylabel("Valor")
plt.legend()
plt.show()
```

```python
# Ejercicio 39
from scipy.integrate import odeint

# Simular la propagación de una epidemia en una red usando ecuaciones diferenciales
def epidemic_spread_model(state, t, infection_rate, recovery_rate):
    # 'state' representa la fracción de la población susceptible (S), infectada (I) y recuperada (R)
    S, I, R = state

    # Modelo SIR simplificado para la propagación de la epidemia
    dSdt = -infection_rate * S * I
    dIdt = infection_rate * S * I - recovery_rate * I
    dRdt = recovery_rate * I
    return [dSdt, dIdt, dRdt]

# Parámetros del modelo de propagación
infection_rate_epidemic = 0.3  # Tasa de infección
recovery_rate_epidemic = 0.1  # Tasa de recuperación

# Estado inicial de la población: la mayoría susceptible, algunos infectados, ninguno recuperado
initial_state_epidemic = [
    0.99,
    0.01,
    0.0,
]  # 99% susceptible, 1% infectado, 0% recuperado

# Resolver la ecuación diferencial del modelo de propagación de epidemias
epidemic_spread_solution = odeint(epidemic_spread_model, initial_state_epidemic, time,
    args=(infection_rate_epidemic, recovery_rate_epidemic))

# Visualización de la propagación de la epidemia
plt.plot(time, epidemic_spread_solution[:, 0], label="Susceptible")
plt.plot(time, epidemic_spread_solution[:, 1], label="Infectado")
plt.plot(time, epidemic_spread_solution[:, 2], label="Recuperado")
plt.title("Modelado de Propagación de Epidemias en Redes")
plt.xlabel("Tiempo")
plt.ylabel("Fracción de la Población")
plt.legend()
plt.show()
```
