```python
# Ejercicio 35
from datetime import date

# Pedir la fecha de nacimiento
anio_nacimiento = int(input("Ingrese su año de nacimiento: "))
mes_nacimiento = int(input("Ingrese su mes de nacimiento (1-12): "))
dia_nacimiento = int(input("Ingrese su día de nacimiento: "))

# Obtener la fecha actual
hoy = date.today()
fecha_nacimiento = date(anio_nacimiento, mes_nacimiento, dia_nacimiento)

# Calcular la diferencia en años, meses y días
edad_anios = hoy.year - fecha_nacimiento.year
edad_meses = hoy.month - fecha_nacimiento.month
edad_dias = hoy.day - fecha_nacimiento.day

# Ajustar si el mes aún no ha pasado
if edad_meses < 0 or (edad_meses == 0 and edad_dias < 0):
    edad_anios -= 1
    edad_meses += 12

# Ajustar si el día aún no ha pasado
if edad_dias < 0:
    from calendar import monthrange

    mes_anterior = hoy.month - 1 if hoy.month > 1 else 12
    anio_anterior = hoy.year if hoy.month > 1 else hoy.year - 1
    dias_mes_anterior = monthrange(anio_anterior, mes_anterior)[1]
    edad_meses -= 1
    edad_dias += dias_mes_anterior

print(f"Tienes {edad_anios} años, {edad_meses} meses y {edad_dias} días.")
```

```python
# Ejercicio 112
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(A | B)  # Unión: {1, 2, 3, 4, 5, 6}
print(A & B)  # Intersección: {3, 4}
print(A - B)  # Diferencia: {1, 2}
print(A ^ B)  # Diferencia simétrica: {1, 2, 5, 6}
```
