```python
# Ejercicio 1
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import warnings
warnings.filterwarnings("ignore")

# Distribución Binomial
# Un ejemplo práctico es el cálculo de la probabilidad de obtener un número determinado de caras al lanzar una moneda.
# Parámetros de la distribución binomial
n = 50  # número de lanzamientos de la moneda
p = 0.5  # probabilidad de obtener cara en un lanzamiento

# Generar una distribución binomial
binomial_dist = stats.binom(n, p)
x = np.arange(0, n + 1)

# Calcular la PMF (Función de Masa de Probabilidad)
pmf = binomial_dist.pmf(x)

# Graficar la distribución
plt.figure(figsize=(10, 6))
plt.bar(x, pmf, color="skyblue")
plt.title("Distribución Binomial de 10 Lanzamientos de una Moneda")
plt.xlabel("Número de Caras Obtenidas")
plt.ylabel("Probabilidad")
plt.xticks(x)
plt.grid(axis="y", alpha=0.75)
plt.show()
```

```python
# Ejercicio 2
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn3
import numpy as np

# Supongamos tres eventos con las siguientes probabilidades:
P_A = 0.30  # Probabilidad de A
P_B = 0.25  # Probabilidad de B
P_C = 0.45  # Probabilidad de C

# Probabilidades de intersección (para simplificar el ejemplo, usaremos probabilidades inventadas)
P_A_and_B = 0.05  # Probabilidad de A y B
P_A_and_C = 0.04  # Probabilidad de A y C
P_B_and_C = 0.04  # Probabilidad de B y C
P_A_and_B_and_C = 0.02  # Probabilidad de A, B y C

# Dibujar el diagrama de Venn
plt.figure(figsize=(8, 8))
venn = venn3(
    subsets=(P_A, P_B, P_A_and_B, P_C, P_A_and_C, P_B_and_C, P_A_and_B_and_C),
    set_labels=("A", "B", "C"),
)

# Asignar colores
for subset in ["100", "010", "110", "001", "101", "011", "111"]:
    venn.get_patch_by_id(subset).set_color(
        np.random.rand(
            3,
        )
    )

# Añadir las probabilidades a cada intersección
venn.get_label_by_id("100").set_text(
    f"{P_A - P_A_and_B - P_A_and_C + P_A_and_B_and_C:.2f}"
)
venn.get_label_by_id("010").set_text(
    f"{P_B - P_A_and_B - P_B_and_C + P_A_and_B_and_C:.2f}"
)
venn.get_label_by_id("001").set_text(
    f"{P_C - P_A_and_C - P_B_and_C + P_A_and_B_and_C:.2f}"
)
venn.get_label_by_id("110").set_text(f"{P_A_and_B - P_A_and_B_and_C:.2f}")
venn.get_label_by_id("101").set_text(f"{P_A_and_C - P_A_and_B_and_C:.2f}")
venn.get_label_by_id("011").set_text(f"{P_B_and_C - P_A_and_B_and_C:.2f}")
venn.get_label_by_id("111").set_text(f"{P_A_and_B_and_C:.2f}")
plt.title("Diagrama de Venn para Tres Eventos con Probabilidades")
plt.show()
```

```python
# Ejercicio 8
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Probabilidades iniciales
P_C1 = 0.4  # Probabilidad de que un dato pertenezca a la clase C1
P_C2 = 0.6  # Probabilidad de que un dato pertenezca a la clase C2

# Probabilidades condicionales de observar cierta característica (X) dado cada clase
P_X_given_C1 = 0.7
P_X_given_C2 = 0.2

# Probabilidad de observar la característica (X) en general
P_X = P_X_given_C1 * P_C1 + P_X_given_C2 * P_C2

# Aplicación del Teorema de Bayes para calcular la probabilidad posterior
# de que el dato pertenezca a C1 o C2 dado que observamos la característica X
P_C1_given_X = (P_X_given_C1 * P_C1) / P_X
P_C2_given_X = (P_X_given_C2 * P_C2) / P_X

# Visualización
labels = ["P(C1|X)", "P(C2|X)"]
posterior_probabilities = [P_C1_given_X, P_C2_given_X]
plt.figure(figsize=(6, 4))
plt.bar(labels, posterior_probabilities, color=["blue", "green"])
plt.title("Probabilidades Posteriores con el Teorema de Bayes")
plt.ylabel("Probabilidad")
plt.ylim(0, 1)
plt.show()
print(f"Probabilidad de C1 dado X: {P_C1_given_X:.2f}")
print(f"Probabilidad de C2 dado X: {P_C2_given_X:.2f}")
```

```python
# Ejercicio 9
import numpy as np
import matplotlib.pyplot as plt

# Generar muestras de diferentes tamaños de una distribución normal
mu = 0  # Media de la distribución
sigma = 1  # Desviación estándar de la distribución
sample_sizes = [10, 100, 1000, 10000, 100000, 1000000]  # Diferentes tamaños de muestra

# Almacenar las medias de las muestras
sample_means = []

# Calcular la media de las muestras para cada tamaño de muestra
for size in sample_sizes:
    sample = np.random.normal(mu, sigma, size)
    sample_means.append(np.mean(sample))

# Graficar cómo la media de las muestras converge hacia la media de la distribución
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, sample_means, marker="o", linestyle="-", color="b")
plt.axhline(y=mu, color="r", linestyle="--", label="Media Teórica")
plt.title("Ley de los Grandes Números: Convergencia de la Media de la Muestra")
plt.xscale("log")
plt.xlabel("Tamaño de la Muestra (Escala Logarítmica)")
plt.ylabel("Media de la Muestra")
plt.legend()
plt.show()
```

```python
# Ejercicio 11
import numpy as np
import matplotlib.pyplot as plt

def convex_function(x):
    return x**2

# Generar valores aleatorios de una variable aleatoria
np.random.seed(0)  # Para resultados reproducibles
random_values = np.random.normal(0, 1, 1000)  # 1000 muestras de una distribución normal

# Calcular la esperanza de la función convexa (media de f(x))
expected_value_of_function = np.mean(convex_function(random_values))

# Calcular la función de la esperanza (f(media de x))
function_of_expected_value = convex_function(np.mean(random_values))

# Graficar la función y los puntos
x = np.linspace(-3, 3, 1000)
y = convex_function(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="f(x) = x^2", color="blue")
plt.scatter(np.mean(random_values), expected_value_of_function, color="red",
    label="Esperanza de f(x)")
plt.scatter( np.mean(random_values), function_of_expected_value, color="black",
    label="f(Esperanza de x)")
plt.axvline(x=np.mean(random_values), color="gray", linestyle="--")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Desigualdad de Jensen")
plt.legend()
plt.show()
print(f"Esperanza de f(x): {expected_value_of_function:.2f}")
print(f"f(Esperanza de x): {function_of_expected_value:.2f}")
```

```python
# Ejercicio 12
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra de una variable aleatoria no negativa
# Usaremos una distribución exponencial como ejemplo
lambda_param = 1.0  # Parámetro lambda de la distribución exponencial
sample_size = 10000
samples = np.random.exponential(1 / lambda_param, sample_size)

# Calcular el límite para aplicar la Desigualdad de Markov
bound = 3  # Valor para el cual queremos calcular la probabilidad
mean_sample = np.mean(samples)
markov_bound = mean_sample / bound

# Calcular la probabilidad real para comparar
real_probability = np.mean(samples > bound)

# Graficar la distribución y los límites
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=30, density=True, color="skyblue", alpha=0.7, label="Distribución de la Muestra")
plt.axvline(x=bound, color="red", linestyle="--", label=f"Límite ({bound})")
plt.axhspan(
    0,
    markov_bound,
    xmin=0,
    xmax=bound / np.max(samples),
    alpha=0.3,
    color="orange",
    label="Área bajo Markov Bound",
)
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.title("Ejemplo de la Desigualdad de Markov")
plt.legend()
plt.show()
print(f"Media de la muestra: {mean_sample:.2f}")
print(f"Cota superior de Markov: {markov_bound:.2f}")
print(
    f"Probabilidad real de que la muestra sea mayor que {bound}: {real_probability:.2f}"
)
```

```python
# Ejercicio 13
import numpy as np
import matplotlib.pyplot as plt

# Supongamos que tenemos dos eventos condicionales B1 y B2
# y queremos calcular la probabilidad de un evento A.
# Probabilidades marginales de B1 y B2
P_B1 = 0.6
P_B2 = 0.4

# Probabilidades condicionales de A dado B1 y B2
P_A_given_B1 = 0.7
P_A_given_B2 = 0.3

# Aplicación del Teorema de Probabilidad Total
# P(A) = P(A|B1)P(B1) + P(A|B2)P(B2)
P_A = P_A_given_B1 * P_B1 + P_A_given_B2 * P_B2

# Visualización
labels = ["P(A|B1)P(B1)", "P(A|B2)P(B2)", "P(A)"]
values = [P_A_given_B1 * P_B1, P_A_given_B2 * P_B2, P_A]
plt.figure(figsize=(8, 5))
plt.bar(labels, values, color=["blue", "green", "red"])
plt.title("Teorema de Probabilidad Total")
plt.ylabel("Probabilidad")
plt.ylim(0, 1)
plt.show()
print(f"Probabilidad Total de A: P(A) = {P_A:.2f}")
```

```python
# Ejercicio 14
import numpy as np
import matplotlib.pyplot as plt

# Supongamos tres eventos: A, B y C
# Queremos calcular la probabilidad conjunta P(A, B, C) utilizando la regla de la cadena.
# Probabilidades iniciales
P_A = 0.6  # Probabilidad de A
P_B_given_A = 0.7  # Probabilidad de B dado A
P_C_given_A_and_B = 0.8  # Probabilidad de C dado A y B

# Aplicación del Teorema de la Regla de la Cadena
# P(A, B, C) = P(C|A, B) * P(B|A) * P(A)
P_A_and_B_and_C = P_C_given_A_and_B * P_B_given_A * P_A

# Visualización
events = ["P(A)", "P(B|A)", "P(C|A, B)", "P(A, B, C)"]
probabilities = [P_A, P_B_given_A, P_C_given_A_and_B, P_A_and_B_and_C]
plt.figure(figsize=(8, 5))
plt.bar(events, probabilities, color=["blue", "green", "red", "purple"])
plt.title("Teorema de la Regla de la Cadena en Probabilidad Conjunta")
plt.ylabel("Probabilidad")
plt.ylim(0, 1)
plt.show()
print(f"Probabilidad Conjunta P(A, B, C): {P_A_and_B_and_C:.2f}")
```

```python
# Ejercicio 15
import numpy as np
import matplotlib.pyplot as plt

# Definir dos distribuciones independientes: distribución normal y distribución exponencial
mu, sigma = 0, 1  # Parámetros de la distribución normal
lambda_exp = 1  # Parámetro de la distribución exponencial

# Tamaño de la muestra
sample_size = 10000

# Generar muestras de las dos distribuciones
normal_samples = np.random.normal(mu, sigma, sample_size)
exponential_samples = np.random.exponential(1 / lambda_exp, sample_size)

# Sumar las dos series de muestras para obtener la convolución
sum_samples = normal_samples + exponential_samples

# Graficar las distribuciones individuales y la suma (convolución)
plt.figure(figsize=(12, 6))

# Distribución normal
plt.subplot(1, 3, 1)
plt.hist(normal_samples, bins=30, density=True, alpha=0.6, color="blue")
plt.title("Distribución Normal")
plt.xlabel("Valor")
plt.ylabel("Densidad")

# Distribución exponencial
plt.subplot(1, 3, 2)
plt.hist(exponential_samples, bins=30, density=True, alpha=0.6, color="green")
plt.title("Distribución Exponencial")
plt.xlabel("Valor")

# Convolución de las dos distribuciones
plt.subplot(1, 3, 3)
plt.hist(sum_samples, bins=30, density=True, alpha=0.6, color="red")
plt.title("Convolución de Normal y Exponencial")
plt.xlabel("Valor")
plt.tight_layout()
plt.show()
```

```python
# Ejercicio 37
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generar datos aleatorios para dos muestras
datos1 = np.random.normal(
    0, 1, 200
)  # Muestra 1: distribución normal, media 0, desviación estándar 1
datos2 = np.random.normal(
    0.5, 1, 200
)  # Muestra 2: distribución normal, media 0.5, desviación estándar 1

# Realizar una prueba t de Student
t_stat, p_value = stats.ttest_ind(datos1, datos2)

# Nivel de significancia
nivel_significancia = 0.05

# Creación de un gráfico
plt.figure(figsize=(8, 5))
plt.hist(datos1, bins=30, color="blue", alpha=0.5, label="Muestra 1")
plt.hist(datos2, bins=30, color="green", alpha=0.5, label="Muestra 2")
plt.axvline(np.mean(datos1), color="blue", linestyle="dashed", label="Media Muestra 1")
plt.axvline(np.mean(datos2), color="green", linestyle="dashed", label="Media Muestra 2")
plt.title(f"Prueba t de Student\nEstadístico t: {t_stat:.2f}, p-valor: {p_value:.3f}")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.legend()

# Agregar anotaciones sobre la interpretación
if p_value < nivel_significancia:
    plt.annotate("Rechazar hipótesis nula", (0.25, 25), fontsize=12, color="red")
else:
    plt.annotate("No rechazar hipótesis nula", (0.25, 25), fontsize=12, color="green")
plt.show()
```

```python
# Ejercicio 38
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from scipy import stats

# Generar datos aleatorios para simular un conjunto de datos de clasificación
X = np.random.rand(300, 5)  # 300 observaciones con 5 características
y = np.random.randint(0, 2, 300)  # Etiquetas de clase binaria

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Entrenar un modelo de clasificación (ejemplo: RandomForest)
modelo = RandomForestClassifier()
modelo.fit(X_train, y_train)

# Predecir las etiquetas de los datos de prueba
y_pred = modelo.predict(X_test)

# Calcular la precisión del modelo
precision = accuracy_score(y_test, y_pred)

# Realizar una prueba t de Student para comparar la precisión con un umbral deseado (ejemplo: 0.75)
t_stat, p_valor = stats.ttest_1samp(y_pred, 0.75)

# Visualización
plt.figure(figsize=(8, 5))
plt.hist(y_pred, bins=10, color="blue", alpha=0.7)
plt.axvline(x=0.75, color="red", linestyle="--", label="Umbral deseado")
plt.title(
    f"Precisión del Modelo: {precision:.2f}, Estadístico t: {t_stat:.2f}, p-valor: {p_valor:.3f}"
)
plt.xlabel("Predicciones")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```

```python
# Ejercicio 43
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Generar dos conjuntos de datos aleatorios (muestras) para el análisis
datos1 = np.random.normal(
    0, 1, 700
)  # Muestra 1: distribución normal, media 0, desviación estándar 1
datos2 = np.random.normal(
    0.5, 1, 700
)  # Muestra 2: distribución normal, media 0.5, desviación estándar 1

# Realizar una prueba t de Student para comparar las medias de dos muestras independientes
t_stat, p_value = stats.ttest_ind(datos1, datos2)

# Crear un DataFrame para visualización con Seaborn
df1 = pd.DataFrame({"Datos": datos1, "Grupo": "Grupo 1"})
df2 = pd.DataFrame({"Datos": datos2, "Grupo": "Grupo 2"})
df = pd.concat([df1, df2])

# Configurar el estilo de Seaborn y crear un gráfico de caja
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(x="Grupo", y="Datos", data=df)
plt.title(f"Prueba t de Student\nEstadístico t: {t_stat:.2f}, p-valor: {p_value:.3f}")
plt.show()
```

```python
# Ejercicio 44
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway

# Generar dos conjuntos de datos aleatorios (muestras) con diferentes varianzas
datos1 = np.random.normal(
    0, 1, 700
)  # Muestra 1: distribución normal, media 0, desviación estándar 1
datos2 = np.random.normal(
    0, 2, 700
)  # Muestra 2: distribución normal, media 0, desviación estándar 2

# Realizar una prueba F de Fisher
f_stat, p_value = f_oneway(datos1, datos2)

# Crear un DataFrame para visualización
df = pd.DataFrame({"Grupo 1": datos1, "Grupo 2": datos2})

# Configurar el estilo de Seaborn y crear un gráfico de densidad
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, fill=True)
plt.title(f"Prueba F de Fisher\nEstadístico F: {f_stat:.2f}, p-valor: {p_value:.3f}")
plt.show()
```
