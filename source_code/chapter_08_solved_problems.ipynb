```python
# Ejercicio 16
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(42)
X_train = np.random.rand(100, 1)
y_train = 2 * X_train + 1 + 0.1 * np.random.randn(100, 1)

# Definir el modelo lineal
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])

# Compilar el modelo
model.compile(optimizer="sgd", loss="mean_squared_error")

# Entrenar el modelo
history = model.fit(X_train, y_train, epochs=50, verbose=0)

# Visualizar el proceso de entrenamiento
plt.plot(history.history["loss"])
plt.title("Model Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()

# Hacer predicciones con el modelo entrenado
X_test = np.array([[0.2], [0.5], [0.8]])
predictions = model.predict(X_test)

# Mostrar las predicciones
print("\nPredicciones del modelo:")
for i in range(len(X_test)):
    print(f"Entrada: {X_test[i][0]}, Predicción: {predictions[i][0]}")
```

```python
# Ejercicio 17
def ordenamiento_burbuja(lista):
    n = len(lista)
    for i in range(n):
        for j in range(0, n - i - 1):
            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
    return lista


print(ordenamiento_burbuja([64, 34, 25, 12, 22, 11, 90]))
```

```python
# Ejercicio 20
import torch
import torch.nn as nn
import torch.nn.functional as F


class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, 3, 1)
        self.conv2 = nn.Conv2d(32, 64, 3, 1)
        self.fc1 = nn.Linear(9216, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.max_pool2d(x, 2)
        x = torch.flatten(x, 1)  # Asegúrate de tener 'torch.' antes de flatten
        x = F.relu(self.fc1(x))
        output = self.fc2(x)
        return output


modelo = ConvNet()
print(modelo)
```
